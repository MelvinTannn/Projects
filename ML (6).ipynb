```python
import pandas as pd
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
data
```

```python
x_data = data[['Age','EstimatedSalary']]
y_data = data[['Purchased']]
```

Pada code diatas diambil fitur berupa kolom age dan estimated salary dengan hasil kelas klassifikasi pada kolom purchased

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=0)
```

Pada code diatas dataset displit menjadi 0.8 train dan 0.2 test dengan randomstate = 0 yang berarti data tidak dishuffle. karena menurut saya data awalnya sudah cukup acak.

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=3)

model.fit(X_train, Y_train)
```

Pada code diatas data train dimasukkan ke dalam model yaitu decision tree classifier

```python
import matplotlib.pyplot as plt
from sklearn import tree

plt.rcParams['figure.dpi'] = 100
plt.subplots(figsize=(15, 10))
tree.plot_tree(model, fontsize=10)
plt.show()
```

Pada hasil classification tree diatas dapat dilihat bahwa X[0] <= 44.5 pada root itu sendiri berarti bahwa splitter criteria yang diambil adalah pada kelas 0 dengan kondisi jika nilai <= 44.5 maka ke kanan jika tidak ke kiri. untuk gini sendiri berarti nilai gini impurity dimana artinya data yang dibagi dimana jika gini nilainya yaitu 0 maka splitter sudah mendapatkan hasil splitting yang baik atau membagi dengan sempurna ke kelasnya lalu untuk samples sendiri berarti jumlah data yang dimasukkan ke dalam splitter. dan terakhir value [199, 121] artinya jumlah data yang displit oleh splitter ke kedua belah kelas yaitu kelas purchased=1 dan purchased=0

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(Y_test, y_pred))
```

Dari model diatas didapatkan akurasi sebesar 95% atau 0.95
